### MERGING THE DATASETS

```python
# Concatenate the DataFrames along rows 
frames = [ipc_df, women_df, children_df]

# Concatenate them into a single DataFrame
crime = pd.concat(frames, ignore_index=True)

# Display the merged DataFrame to verify
print(crime.head())
```

```python
# Filtering out rows where DISTRICT or STATE/UT is 'TOTAL'
crime = crime[(crime['DISTRICT'] != 'TOTAL') & (crime['STATE/UT'] != 'TOTAL')]

# Reset index if needed
crime.reset_index(drop=True, inplace=True)

# Display first few rows to verify
print(crime.head())
```

```python
crime.shape
```

```python
# Filter out rows where 'DISTRICT' or 'STATE/UT' contains the word 'TOTAL'
crime = crime[~crime['DISTRICT'].str.contains(r'\bTOTAL\b', case=False)]
crime = crime[~crime['STATE/UT'].str.contains(r'\bTOTAL\b', case=False)]

# Reset index if needed
crime.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
print(crime)
```

## 4.1

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Selecting relevant columns for clustering
cols_for_clustering = ['POPULATION 2001', 'POPULATION 2011', 'LITERACY RATE 2001', 'LITERACY RATE 2011', 'AREA km2', 'TOTAL']
```

```python
# Extracting the data for clustering
X = crime[cols_for_clustering]
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Performing K-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
crime['Cluster'] = kmeans.fit_predict(X_scaled)
```

## 4.2

```python
# Creating DataFrames for each cluster
sensitive_df = crime[crime['Cluster'] == 1]
moderate_df = crime[crime['Cluster'] == 0]
peaceful_df = crime[crime['Cluster'] == 2]
```
